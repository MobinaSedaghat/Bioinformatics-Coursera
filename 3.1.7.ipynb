```python
def distance_Between_Pattern_And_Strings(pattern, dna): 
    k = len(pattern)
    distance = 0
    for string in dna:
        hamming_dist = 1000
        for i in range(len(string) - k + 1):
            if hamming_dist > hamming(pattern, string[i:i+k]):
                hamming_dist = hamming(pattern , string[i:i+k])
        distance = distance + hamming_dist
    return distance
```

```python
def hamming(str1, str2):
    dist = 0
    for i in range (len(str1)):
        if str1[i] != str2[i]: 
            dist = dist + 1
    return dist
        
```
